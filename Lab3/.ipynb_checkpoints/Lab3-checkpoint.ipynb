Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Part 1
---

```python
data = np.loadtxt('lab3data.csv', delimiter = ',', skiprows = 1)

pt1 = data[:, 0]
pt2 = data[:, 1]
eta1 = data[:, 2]
eta2 = data[:, 3]
phi1 = data[:, 4]
phi2 = data[:, 5]
E1 = data[:, 6]
E2 = data[:, 7]
```

```python
px = pt1 * np.cos(phi1) + pt2 * np.cos(phi2)
py = pt1 * np.sin(phi1) + pt2 * np.sin(phi2)
pz = pt1 * np.sinh(eta1) + pt2 * np.sinh(eta2)
M = np.sqrt(abs((E1 + E2)**2 - (px**2 + py**2 + pz**2)))
```

```python
bins = np.linspace(80, 100, 41)

counts, bin_edges = np.histogram(M, bins = bins)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

error = np.sqrt(counts)

fig, ax = plt.subplots()
ax.errorbar(bin_centers, counts, yerr = error, fmt = 'o', color = 'k', ecolor = 'r', label = 'data')

ax.set_xlabel("Invariant Mass (GeV)")
ax.set_ylabel('Number of Events')
ax.set_title('Invariant Mass Distribution')
```

# Part 2 #
---

```python
def decay_dist(m, m0, gamma):
    return (1 / np.pi) * ((gamma / 2) / ((m - m0)**2 + (gamma / 2)**2))
```

```python
from scipy.optimize import curve_fit

bins = np.linspace(80, 100, 42)
hist_vals, bin_edges = np.histogram(M, bins=bins)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

mask = (bin_centers > 87) & (bin_centers < 93)
x_fit = bin_centers[mask]
y_fit = hist_vals[mask]

def breit_wigner(m, m0, gamma):
    N = 2500
    return (N / np.pi) * ((gamma / 2) / ((m - m0)**2 + (gamma / 2)**2))

popt, pcov = curve_fit(breit_wigner, x_fit, y_fit, p0 = [91, 2])

m_plot = np.linspace(80, 100, 100)
fit_curve = breit_wigner(m_plot, *popt)

fig, ax = plt.subplots()
ax.hist(M, bins = bins, label = 'Data', color = 'k')
ax.plot(m_plot,fit_curve, label = f'Breit-Wigner Fit', color = 'r')
ax.set_xlabel("Invariant Mass (GeV)")
ax.set_ylabel('Counts per bin')
ax.set_title('Invariant Mass with Breit-Wigner Fit')
ax.legend()
```

```python
m_plot = np.linspace(80, 100, 1000)
y_fit_full = breit_wigner(bin_centers, *popt)

residuals = hist_vals - y_fit_full
errors = np.sqrt(hist_vals)

fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(9, 7), gridspec_kw={'height_ratios': [3, 1]})


ax1.errorbar(bin_centers, hist_vals, yerr=errors, fmt='o', color='black', label='Data', capsize=2)
ax1.plot(m_plot, breit_wigner(m_plot, *popt), color='r', label='Breit-Wigner Fit')

ax1.set_ylabel('Counts per Bin')
ax1.set_title('Invariant Mass Distribution with Breit-Wigner Fit')
ax1.legend()
ax1.grid()

ax2.axhline(0, color='gray', linestyle='--', linewidth=3)
ax2.plot(bin_centers, residuals, 'o', color='b', label='Residuals')
ax2.set_xlabel('Invariant Mass (GeV)')
ax2.set_ylabel('Residuals')
ax2.grid()

ax2.axvline(87, color='k', linestyle=':', linewidth = 3)
ax2.axvline(93, color='k', linestyle=':', linewidth = 3)
```

```python
from scipy.stats import chi2

y_obs = hist_vals[mask]
y_err = np.sqrt(hist_vals[mask])
y_exp = breit_wigner(x_fit, *popt)

chi_squared = np.sum(((y_obs - y_exp)**2) / (y_err**2))

ndof = len(y_obs) - 2

chi_squared_red = chi_squared / ndof

p_value = 1 - chi2.cdf(chi_squared, df=ndof)


print(f'Chi-square: {chi_squared}')
print(f'Reduced Chi-square: {chi_squared_red}')
print(f'Degrees of freedom: {ndof}')
print(f'p-value: {p_value}')
```

```python
m0_best = popt[0]
m0_uncertainty = np.sqrt(pcov[0, 0])

print(f"Best-fit m0 = {m0_best:} +- {m0_uncertainty:} GeV")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(9, 7), gridspec_kw={'height_ratios': [3, 1]})

ax1.errorbar(bin_centers, hist_vals, yerr=errors, fmt='o', color='k', capsize=2, label='Data')
ax1.plot(m_plot, breit_wigner(m_plot, *popt), color='r', label='Breit-Wigner Fit')

ax1.set_ylabel('Counts per Bin')
ax1.set_title('Invariant Mass Distribution with Breit-Wigner Fit')
ax1.legend()
ax1.grid()

annotation_text = (
    f"$m_0$ = {m0_best:.1f} +- {m0_uncertainty:.1f} GeV\n"
    r"$\chi^2$/ndof = {chi_squared_red:.1f}\n"
    f"p-value = {p_value:.1f}"
)
ax1.text(0.97, 0.95, annotation_text, transform=ax1.transAxes,
         fontsize=10, ha='right', va='top', bbox=dict(facecolor='white', edgecolor='gray'))

ax2.axhline(0, color='gray', linestyle='--')
ax2.plot(bin_centers, residuals, 'o', color='b', label='Residuals')
ax2.set_xlabel('Invariant Mass (GeV)')
ax2.set_ylabel('Residuals')
ax2.grid()

for x in [87, 93]:
    ax2.axvline(x, color='black', linestyle=':', linewidth=3)

ax1.set_title("Figure 1", size = 25)
```

# Part 3 #
---

```python
m0_vals = np.linspace(89, 91, 300)
gamma_vals = np.linspace(5, 8, 300)
chi2_grid = np.zeros((len(gamma_vals), len(m0_vals)))  # shape: (gamma, m0)

y_obs = hist_vals[mask]
y_err = np.sqrt(hist_vals[mask])
x_fit = bin_centers[mask]

for i, gamma in enumerate(gamma_vals):
    for j, m0 in enumerate(m0_vals):
        y_model = breit_wigner(x_fit, m0, gamma)
        chi2 = np.sum(((y_obs - y_model)**2) / y_err**2)
        chi2_grid[i, j] = chi2

M0_grid, Gamma_grid = np.meshgrid(m0_vals, gamma_vals)
```

```python
chi2_min = np.min(chi2_grid)
delta_chi2 = chi2_grid - chi2_min
delta_chi2_clipped = np.clip(delta_chi2, 0, 35)

levels_conf = [2.30, 11.83]
styles_conf = ['solid', 'dashed']
labels_conf = {2.30: r'1$\sigma$', 11.83: r'3$\sigma$'}

fig, ax = plt.subplots(figsize=(8, 6))

contourf = ax.contourf(M0_grid, Gamma_grid, delta_chi2_clipped, levels=100, cmap='plasma')
cbar = plt.colorbar(contourf, ax=ax)
cbar.set_label(r'$\Delta \chi^2$', fontsize=12)
cbar.ax.tick_params(labelsize=10)

for lvl, style in zip(levels_conf, styles_conf):
    cs = ax.contour(M0_grid, Gamma_grid, delta_chi2, levels=[lvl],
                    colors='white', linestyles=style, linewidths=1.5)
    ax.clabel(cs, fmt={lvl: labels_conf[lvl]}, colors='white', fontsize=10)

ax.plot(popt[0], popt[1], 'ro', markersize=8, label='Best Fit')
ax.plot(popt[0], popt[1], 'kx', markersize=8, mew=2)

ax.set_xlabel(r'$m_0$ [GeV]', fontsize=12)
ax.set_ylabel(r'$\Gamma$ [GeV]', fontsize=12)
ax.set_title("Figure 2", size=25)
ax.legend()
```
