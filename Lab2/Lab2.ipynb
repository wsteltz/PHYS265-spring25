```python
R = 6378100
g = 9.81
alpha = 0
gamma = 1

def grav(r):
    return g * (r/R)

def p2(t, Y):
    y, v = Y
    r = R + y
    dydt = v
    dvdt = -grav(r) - alpha * abs(v) ** gamma * np.sign(v)
    return [dydt, dvdt]

t_span = (0,100)

t_eval = np.linspace(0, 100, 100)
y0 = 0
v0 = 0

def hit(t, Y):
    y, v = Y
    return y + 4000

hit.terminal = True

soln = solve_ivp(p2, t_span, (y0, v0), t_eval = t_eval, events = hit)

fig, ax1 = plt.subplots()
ax1.plot(soln.t, soln.y[0], 'b')
ax2 = ax1.twinx()
ax2.plot(soln.t, soln.y[1], 'r')
ax1.set_xlabel('Time (s)', size = 15)
ax1.set_ylabel('Position (m)', size = 15, color = 'b')
ax2.set_ylabel('Velocity (m/s)', size = 15, color = 'r')
ax1.set_title('Figure 2', size = 20)
```

Incorporating a height-dependent g will INCREASE the fall time because you are multiplying the original gravity by a number that is less than 1. ((R + y) / R) where y < 0.

```python
v0 = 0
y0 = 6378100
alpha = 0
gamma = 1
g = 9.81
R = 6378100
 
def position(t, Y):
    y, v = Y
    dydt = v
    dvdt = (-g * y/R) - alpha * v ** gamma
    return [dydt, dvdt]

t_span = (0, 5000)
t_eval = np.linspace(0, 5000, 10000)

sol = solve_ivp(position, t_span, (y0, v0), t_eval = t_eval)

t_val = sol.t
y_val = sol.y[0]
v_val = sol.y[0]

cross = np.where(y_val < -R)[0]

if len(cross) > 0:
    idx = cross[0]
    
    t1, t2 = t_val[idx - 1], t_val[idx]
    y1, y2 = y_val[idx - 1], y_val[idx]
    v1, v2 = v_val[idx - 1], v_val[idx]
    
    t_cross = t1 + (t2 - t1) * ((-R - y1) / (y2 - y1))
    v_cross = v1 + (v2 - v1) * ((-R - y1) / (y2 - y1))
    
    print(f'The test mass crosses the other side at t = {t_cross} seconds with a velocity of {v_cross} m/s')

else:
    print('The test mass did not reach the other side')

fig, ax = plt.subplots()
ax.plot(sol.t, sol.y[0], 'b')
ax1 = ax.twinx()
ax1.plot(sol.t, sol.y[1], 'r')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Position (m)', size = 15, color = 'b')
ax1.set_ylabel('Velocity (m/s)', size = 15, color = 'r')
ax.set_title('Figure 4', size = 20)
```

```python
R = 6378100

def density(r, n):
    return (1 - (r**2) / (R**2))**n

r_val = np.linspace(0, R, 1000)

n_val = [0, 1, 2, 9]
color = ['b', 'g', 'r', 'k']

fig, ax = plt.subplots()

for n, color in zip(n_val, color):
    rho_val = density(r_val, n)
    ax.plot(r_val / R, rho_val, label = f'n = {n}', color = color)

ax.set_title('Figure 5', size = 20)
ax.set_xlabel('Radius (r / R) (m)')
ax.set_ylabel('Normalized density')
ax.legend()
```

```python
from scipy.integrate import quad
G = 6.6743e-11
Me = 5.972e24

def density(r, n):
    return (1 - (r**2) / R**2)**n

def rho(n):
    integral, err = quad(lambda r: density(r, n) * r**2, 0, R)
    m_int = 4 * np.pi * integral
    return Me / m_int

def mass(r , n, rho_n):
    integral, err = quad(lambda rp: rho_n * density(rp, n) * rp**2, 0, r)
    return 4 * np.pi * integral

def grav_prof(r_vals, n):
    rho_n = rho(n)
    F_val = np.zeros_like(r_val)
    for i, r in enumerate(r_val):
        if r == 0:
            F_val[i] = 0
        else:
            M_enc = mass(r, n, rho_n)
            F_val[i] = G * M_enc / r**2
    return F_val

r_val = np.linspace(0, R, 501)
n_val = [0, 1, 2, 9]
colors = ['b', 'g', 'r', 'k']

fig, ax = plt.subplots()
for n, color in zip(n_val, colors):
    F_val = grav_prof(r_val, n)
    ax.plot(r_val / R, F_val, label = f'n = {n}', color = color)

ax.set_title('Figure 6', size = 20)
ax.set_xlabel('Radius (r / R)')
ax.set_ylabel('Gravitational Force')
ax.legend()
```

```python
def acc(r, n, rho_n):
    if r == 0:
        return 0
    M_r = mass(r, n, rho_n)
    return -G * M_r / r**2

def motion(t, y, n, rho_n):
    r, v = y
    a = acc(r, n, rho_n)
    return [v, a]

t_span = 3000
t_eval = np.linspace(0, t_span, 1001)
y0 = [R, 0]

fig, ax = plt.subplots()
fig, ax1 = plt.subplots()

for n, color in zip(n_val, colors):
    rho_n = rho(n)
    sol = solve_ivp(motion, [0, t_span], y0, args = (n, rho_n), t_eval = t_eval)
    r_val = sol.y[0]
    v_val = sol.y[1]
    ax.plot(sol.t, r_val / R, label = f'n = {n}', color = color)
    ax1.plot(sol.t, v_val, label = f'n = {n}', color = color)

ax1.set_title('Figure 7', size = 20)
ax.set_title('Figure 7', size = 20)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Position (r / R)')
ax1.set_title('Figure 7', size=20)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Velocity (m/s)')
ax.legend()
ax1.legend()
```

```python
def fall_to_center(n):
    rho_n = rho(n)
    y0 = [R, 0]
    t_span = [0, 3000]

    def event_center(t, y, n, rho_n):
        return y[0]
    event_center.terminal = True
    event_center.direction = -1

    sol = solve_ivp(motion, t_span, y0, args = (n, rho_n), events=event_center)

    time_to_center = sol.t_events[0][0]
    speed_at_center = abs(sol.y_events[0][0][1])

    return time_to_center, speed_at_center

n_values = [0, 1, 2, 9]
print("n\tTime to center (s)\tSpeed at center (m/s)")
print("-" * 45)
for n in n_values:
    t_center, v_center = fall_to_center(n)
    print(f"{n}\t{t_center:.2f}\t\t\t{v_center:.2f}")
```

```python
R_moon = 1738100
M_moon = 7.35e22
G = 6.6743e-11

t_moon = (np.pi / 2) * np.sqrt(R_moon**3 / (G * M_moon))

print(f"Fall time to center of the Moon: {t_moon} seconds")
```

```python
#density = mass / volume
#volume of a sphere = 4/3 * pi * R**3
rho_moon = M_moon / ((4 / 3) * (np.pi) * (R_moon**3))
rho_earth = Me / ((4 / 3) * (np.pi) * (R**3))
print(f' The density of the Moon is {rho_moon} kg/m cubed, while the density of the Earth is {rho_earth} kg/m cubed')
```
